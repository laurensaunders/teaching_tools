```python
import matplotlib.pyplot as plt
import numpy as np
import scipy as sp
import pickle
```

```python
# Start by making a random sample of points
random = np.random.random_sample(20,)
xes = np.linspace(1,20,20)
```

```python
# Create a basic plot
plt.figure(figsize=(6,6))
plt.plot(xes, random, 'C0.', label='Random sample')
plt.legend()
plt.title('Random Sample')
plt.ylabel('Generated random number')
plt.xlabel('Output order number')
plt.show()
```

```python
# Clear the plot so that you can plot something new
plt.clf()
```

```python
# Pro tip: use LaTeX notation in your labels to get symbols
plt.figure(figsize=(6,6))
plt.plot(xes, np.linspace(200,400,20), 'C0.', label='Thermometer 1')
plt.legend()
plt.title('Data points')
plt.ylabel('Temperature ($\mu$K)')
plt.xlabel('Resistance (m$\Omega$)')
plt.show()
```

```python
# Load in a dictionary of data (this format protects you from accidentally overwriting or corrupting your data)
data = pickle.load(open('pyplot_practice_data.pkl','rb'))
print(data.keys())
print(data['d050'].keys())
print(data['d050']['data'].keys())
```

```python
plt.clf()
# plot data from one dictionary entry
plt.figure(figsize=(6.5,6.5))
plt.plot(data['d050']['data']['temps'], data['d050']['data']['volts'], 'k-', label='Raw Data')
plt.plot(data['d050']['calpoints']['temps'], data['d050']['calpoints']['volts'], 'r.', alpha=0.5, label='Calibration Points')
plt.grid()
plt.legend()
plt.title('D-050 Calibration Curve')
plt.ylabel('Voltage (V)')
plt.xlabel('Temperature (T)')
#plt.xlim(1,2); plt.ylim(1.6, 1.7)
plt.show()
```

```python
plt.clf()
# plot all of them on the same plot -- a mess!
plt.figure(figsize=(6.5,6.5))
for diode in data:
    plt.plot(data[diode]['data']['temps'], data[diode]['data']['volts'], '.', alpha=0.2, label=str(diode))
plt.legend()
plt.grid()
plt.title('All of the diodes')
plt.ylabel('Voltage (V)')
plt.xlabel('Temperature (K)')
plt.xlim(124.75,125); plt.ylim(0.938, 0.941)
plt.show()
# notice: only 10 colors in the standard set
```

```python
# Try making subplots instead
plt.clf()

fig = plt.figure(figsize=(20,7))

ax1 = fig.add_subplot(1,3,1)
ax1.plot(data['d050']['data']['temps'], data['d050']['data']['volts'], 'C0-', label='Data')
ax1.plot(data['d050']['calpoints']['temps'], data['d050']['calpoints']['volts'], 'C0.', alpha=0.3, label='Calibration Points')
plt.grid()
plt.legend()
plt.xlabel('Temperature (K)')
plt.ylabel('Voltage (V)')
plt.title('D-050')

ax2 = fig.add_subplot(1,3,2)
ax2.plot(data['d026']['data']['temps'], data['d026']['data']['volts'], 'C1-', label='Data')
ax2.plot(data['d026']['calpoints']['temps'], data['d026']['calpoints']['volts'], 'C1.', alpha=0.3, label='Calibration Points')
plt.grid()
plt.legend()
plt.xlabel('Temperature (K)')
plt.ylabel('Voltage (V)')
plt.title('D-026')

ax3 = fig.add_subplot(1,3,3)
ax3.plot(data['d046']['data']['temps'], data['d046']['data']['volts'], 'C2-', label='Data')
ax3.plot(data['d046']['calpoints']['temps'], data['d046']['calpoints']['volts'], 'C2.', alpha=0.3, label='Calibration Points')
plt.grid()
plt.legend()
plt.xlabel('Temperature (K)')
plt.ylabel('Voltage (V)')
plt.title('D-046')

plt.suptitle('Calibrations for D-050, D-026, D-046')
plt.savefig('three_diodes.png')
plt.show()
```
